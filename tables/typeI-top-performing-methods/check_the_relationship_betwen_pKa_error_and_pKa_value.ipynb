```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("typeIII_submission_collection.csv", index_col=0)
df["$\Delta$pKa absolute error (calc - exp)"]=np.NaN
for i,row in enumerate(df.iterrows()):
    error = row[1]['$\Delta$pKa error (calc - exp)']
    abs_error = abs(error)
    df.loc[i,"$\Delta$pKa absolute error (calc - exp)"] = abs_error
df.head()
```

```python
df.columns
```

```python
exp_pKas = df['pKa (exp)']

pKa_range_list = ["[2,4)", "[4,6)", "[6,8)", "[8,10)", "[10,12)"]

df_2_4 = df[(2 <= df['pKa (exp)']) & (df['pKa (exp)'] < 4)]
df_4_6 = df[(4 <= df['pKa (exp)']) & (df['pKa (exp)'] < 6)]
df_6_8 = df[(6 <= df['pKa (exp)']) & (df['pKa (exp)'] < 8)]
df_8_10 = df[(8 <= df['pKa (exp)']) & (df['pKa (exp)'] < 10)]
df_10_12 = df[(8 <= df['pKa (exp)']) & (df['pKa (exp)'] < 10)]

abs_error_2_4 = df_2_4['$\Delta$pKa absolute error (calc - exp)']
abs_error_4_6 = df_4_6['$\Delta$pKa absolute error (calc - exp)']
abs_error_6_8 = df_6_8['$\Delta$pKa absolute error (calc - exp)']
abs_error_8_10 = df_8_10['$\Delta$pKa absolute error (calc - exp)']
abs_error_10_12 = df_10_12['$\Delta$pKa absolute error (calc - exp)']

data = [abs_error_2_4 , abs_error_4_6 , abs_error_6_8 , abs_error_8_10 , abs_error_10_12]
```

```python
plt.style.use(["seaborn-talk", "seaborn-whitegrid"])
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 16
plt.tight_layout()
    
fig = plt.figure(figsize =(10, 7)) 
ax = fig.add_subplot(111) 



bp = ax.boxplot(data, patch_artist = False, notch ='True', vert = 1) 

# x-axis labels 
ax.set_xticklabels(pKa_range_list) 

# changing color and linewidth of  medians 
for median in bp['medians']: 
    median.set(color ='black', 
               linewidth = 3) 

ax.set_xlabel("Experimental pKa range")
ax.set_ylabel("|$\Delta$pKa (calc - exp)|")

plt.savefig("absolute_error_in_macro_pKa_predictions_vs_exp_pKa.pdf")
```
