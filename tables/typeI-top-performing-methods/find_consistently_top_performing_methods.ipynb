```python
import pandas as pd
import numpy as np
```

```python
# Microstate matching statistictis or typeI analysis of 8 mol set
df_stats = pd.read_csv("statistics_with_unmatched_pKa_numbers.csv")
df_dom_ms_stats = pd.read_csv("dominant_microstate_statistics.csv")
```

```python
df_stats.head()
```

```python
df_dom_ms_stats.head()
```

```python
# Add dominant microstate accuracy column to statistics
df_stats["dominant microstate accuracy"]=np.NaN

for i, row in enumerate(df_stats.iterrows()):
    receipt_id = row[1]["ID"]
    #print(receipt_id)
    accuracy = df_dom_ms_stats[df_dom_ms_stats["receipt_id"] == receipt_id]["Accuracy"].values[0]
    accuracy_lower_bound = df_dom_ms_stats[df_dom_ms_stats["receipt_id"] == receipt_id]["Accuracy lower bound"].values[0]
    accuracy_upper_bound = df_dom_ms_stats[df_dom_ms_stats["receipt_id"] == receipt_id]["Accuracy upper bound"].values[0]
    #print(accuracy)
    df_stats.loc[i, "dominant microstate accuracy"] = accuracy
    df_stats.loc[i, "dominant_microstate_accuracy_lower_bound"] = accuracy_lower_bound
    df_stats.loc[i, "dominant_microstate_accuracy_upper_bound"] = accuracy_upper_bound
    
df_stats.head()
```

```python
# Calculate sum of unmatched exp pKas and unmatched ored pKas [2,12]
df_stats["missing or extra pKa predictions"] = np.NaN

for i, row in enumerate(df_stats.iterrows()):
    missing_pKa_count = row[1]["unmatched exp pKas"]
    extra_pKa_count = row[1]["unmatched pred pKas [2,12]"]
    missing_or_extra_pKa_count = missing_pKa_count + extra_pKa_count
    df_stats.loc[i, "missing or extra pKa predictions"] = int(missing_or_extra_pKa_count)
    
df_stats.head()
```

## Methods that consistently appear in top 10 

```python
# Create a set of submission IDs of top 10 methods according to RMSE
# Statistics table was already ordered by RMSE
top_N=10

# Order by increasing RMSE
df_stats_rmse =df_stats.sort_values(by='RMSE', ascending=True).reset_index(drop=True)

df_rmse_top = df_stats_rmse.loc[0:top_N,:]
submission_ID_rmse_top = set(df_rmse_top["ID"])
submission_ID_rmse_top
```

```python
# Create a set of submission IDs of top 10 methods according to MAE
df_stats_mae =df_stats.sort_values(by='MAE', ascending=True).reset_index(drop=True)

df_mae_top = df_stats_mae.loc[0:top_N,:]
submission_ID_mae_top = set(df_mae_top["ID"])
submission_ID_mae_top
```

```python
# Create a set of submission IDs of top 10 methods according to R-squareed
df_stats_r2 =df_stats.sort_values(by='R2', ascending=False).reset_index(drop=True)

df_r2_top = df_stats_r2.loc[0:top_N,:]
submission_ID_r2_top = set(df_r2_top["ID"])
submission_ID_r2_top
```

```python
# Create a set of submission IDs of top 10 methods according to Kendall's Tau
df_stats_tau =df_stats.sort_values(by='kendall_tau', ascending=False).reset_index(drop=True)

df_tau_top = df_stats_tau.loc[0:top_N,:]
submission_ID_tau_top = set(df_tau_top["ID"])
submission_ID_tau_top
```

```python
# Create a set of submission IDs of perfect dominant microstate accuacy
df_dmsa_perfect = df_stats[df_stats["dominant microstate accuracy"] == 1.0].reset_index(drop=True)

submission_ID_dmsa_perfect = set(df_dmsa_perfect.ID.values)
submission_ID_dmsa_perfect
```

```python
# Determine intersection of sets: Methods that are in top N according to all 4 statistics
common_methods = (submission_ID_rmse_top & 
                          submission_ID_mae_top & 
                          submission_ID_dmsa_perfect)
print(common_methods)

df_best = df_stats[df_stats["ID"].isin(common_methods)].reset_index(drop=True)
df_best
```

```python
df_best = df_best.drop(columns = ["unmatched pred pKas", "unmatched pred pKas [4,10]", "missing or extra pKa predictions" ])
df_best.to_csv("statistics_of_consistantly_top_performing_methods.csv", index=False)
```

```python
# Organize a CSV table ready for manuscript
df_best = df_best.round(2)
df_best_table = df_best.loc[:,["ID", "name"]]
df_best_table["RMSE"] = np.NaN
df_best_table["MAE"] = np.NaN
df_best_table["R2"] = np.NaN
df_best_table["kendall_tau"] = np.NaN
df_best_table["Unmatched Exp. pKa Count"] = np.NaN
df_best_table["Unmatched Pred. pKa Count [2,12]"] = np.NaN

df_best_table
```

```python
# Write stats with mean [CI] format
for i, row in enumerate(df_best.iterrows()):
    dmsa_str = str(row[1]["dominant microstate accuracy"]) + " [" + str(row[1]["dominant_microstate_accuracy_lower_bound"]) + ", " + str(row[1]["dominant_microstate_accuracy_upper_bound"]) + "]"
    df_best_table.loc[i, "Dominant Microstate Accuracy"] = dmsa_str 
    RMSE_str = str(row[1]["RMSE"]) + " [" + str(row[1]["RMSE_lower_bound"]) + ", " + str(row[1]["RMSE_upper_bound"]) + "]"
    df_best_table.loc[i, "RMSE"] = RMSE_str 
    MAE_str = str(row[1]["MAE"]) + " [" + str(row[1]["MAE_lower_bound"]) + ", " + str(row[1]["MAE_upper_bound"]) + "]"
    df_best_table.loc[i, "MAE"] = MAE_str 
    R2_str = str(row[1]["R2"]) + " [" + str(row[1]["R2_lower_bound"]) + ", " + str(row[1]["R2_upper_bound"]) + "]"
    df_best_table.loc[i, "R2"] = R2_str 
    TAU_str = str(row[1]["kendall_tau"]) + " [" + str(row[1]["kendall_tau_lower_bound"]) + ", " + str(row[1]["kendall_tau_upper_bound"]) + "]"
    df_best_table.loc[i, "kendall_tau"] = TAU_str 
    
    df_best_table.loc[i, "Unmatched Exp. pKa Count"] = round(row[1]["unmatched exp pKas"]) 
    df_best_table.loc[i, "Unmatched Pred. pKa Count [2,12]"] = round(row[1]["unmatched pred pKas [2,12]"]) 
    
    
# Reorder columns
df_best_table = df_best_table[['ID', 'name',
       'Dominant Microstate Accuracy','RMSE', 'MAE', 'R2', 'kendall_tau',
       'Unmatched Exp. pKa Count', 'Unmatched Pred. pKa Count [2,12]']]

# Save as CSV
df_best_table.to_csv("table_of_consistantly_top_performing_methods.csv", index=False)
```
